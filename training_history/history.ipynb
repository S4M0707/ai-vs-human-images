```python
import json
import matplotlib.pyplot as plt
```

```python
with open('vit_training_history.json') as vit_training_history:
    vit_data = vit_training_history.read()
with open('convnextbase_training_history.json') as convnextbase_training_history:
    convnextbase_data = convnextbase_training_history.read()
with open('efficientnetb0_training_history.json') as efficientnetb0_training_history:
    efficientnetb0_data = efficientnetb0_training_history.read()
with open('resnet_training_history.json') as resnet_training_history:
    resnet_data = resnet_training_history.read()
```

```python
vit = json.loads(vit_data)
convnextbase = json.loads(convnextbase_data)
efficientnetb0 = json.loads(efficientnetb0_data)
resnet50 = json.loads(resnet_data)
```

```python
import matplotlib.pyplot as plt

def plot_loss_curve(epochs, train_loss, val_loss, title='Training and Validation Loss',
                    save_path=None):
    """
    Plots a clean, research-paper-style training vs. validation loss curve.

    Parameters:
        epochs (list or array): Epoch indices
        train_loss (list or array): Training loss values
        val_loss (list or array): Validation loss values
        title (str): Title of the plot
        save_path (str or None): If provided, saves the figure to this path
    """
    plt.rcParams.update({
        "font.family": "serif",
        "font.size": 12,
        "axes.labelsize": 14,
        "axes.titlesize": 16,
        "legend.fontsize": 12,
        "xtick.labelsize": 12,
        "ytick.labelsize": 12
    })

    plt.figure(figsize=(8, 6), dpi=300)
    plt.plot(epochs, train_loss, 'o-', color='royalblue', label='Training Loss', linewidth=2, markersize=5)
    plt.plot(epochs, val_loss, 's--', color='forestgreen', label='Validation Loss', linewidth=2, markersize=5)

    plt.title(title, fontweight='bold')
    plt.xlabel('Epochs', fontweight='bold')
    plt.ylabel('Loss', fontweight='bold')
    plt.legend(loc='upper right', frameon=False)
    plt.grid(True, linestyle='--', linewidth=0.6)
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, bbox_inches='tight')
        print(f"Plot saved to {save_path}")
    else:
        plt.show()

def plot_accuracy_curve(epochs, train_acc, val_acc, title='Training and Validation Accuracy',
                        save_path=None):
    """
    Plots a clean, research-paper-style training vs. validation accuracy curve.

    Parameters:
        epochs (list or array): Epoch indices
        train_acc (list or array): Training accuracy values
        val_acc (list or array): Validation accuracy values
        title (str): Title of the plot
        save_path (str or None): If provided, saves the figure to this path
    """
    plt.rcParams.update({
        "font.family": "serif",
        "font.size": 12,
        "axes.labelsize": 14,
        "axes.titlesize": 16,
        "legend.fontsize": 12,
        "xtick.labelsize": 12,
        "ytick.labelsize": 12
    })

    plt.figure(figsize=(8, 6), dpi=300)
    plt.plot(epochs, train_acc, 'o-', color='dodgerblue', label='Training Accuracy', linewidth=2, markersize=5)
    plt.plot(epochs, val_acc, 's--', color='darkorange', label='Validation Accuracy', linewidth=2, markersize=5)

    plt.title(title, fontweight='bold')
    plt.xlabel('Epochs', fontweight='bold')
    plt.ylabel('Accuracy', fontweight='bold')
    plt.legend(loc='lower right', frameon=False)
    plt.grid(True, linestyle='--', linewidth=0.6)
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, bbox_inches='tight')
        print(f"Plot saved to {save_path}")
    else:
        plt.show()

```

# VIT

```python
epochs = range(1, len(vit['loss']) + 1)
```

```python
plot_loss_curve(epochs, vit['loss'], vit['val_loss'], title = 'ViT Loss')
```

```python
plot_accuracy_curve(epochs, vit['accuracy'], vit['val_accuracy'], title = 'ViT Accuracy')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
model_metrics = {
    "resnet50": {
        "accuracy": 0.9687,
        "precision": 0.9765,
        "recall": 0.9605,
        "f1_score": 0.9684,
        "auc_roc": 0.9958
    },
    "convnextbase": {
        "accuracy": 0.9840,
        "precision": 0.9774,
        "recall": 0.9909,
        "f1_score": 0.9841,
        "auc_roc": 0.9986
    },
    "vit": {
        "accuracy": 0.8652,
        "precision": 0.8615,
        "recall": 0.8702,
        "f1_score": 0.8658,
        "auc_roc": 0.9395
    },
    "efficientnetb0": {
        "accuracy": 0.9196,
        "precision": 0.8690,
        "recall": 0.9882,
        "f1_score": 0.9248,
        "auc_roc": 0.9889
    }
}
```

```python
def plot_model_metric_comparison(metric: str):
    if metric not in ["accuracy", "precision", "recall", "f1_score", "auc_roc"]:
        raise ValueError(f"Invalid metric '{metric}'. Choose from: accuracy, precision, recall, f1_score, auc_roc")
    
    # Prepare dataframe
    df = pd.DataFrame(model_metrics).T.reset_index().rename(columns={'index': 'Model'})
    df['Model'] = df['Model'].str.replace('b0', 'B0') \
                             .str.replace('convnextbase', 'ConvNeXt-Base') \
                             .str.replace('resnet50', 'ResNet-50') \
                             .str.replace('vit', 'ViT')
    
    # Plot settings
    sns.set_context("notebook", font_scale=1.3)
    sns.set_style("whitegrid")
    plt.rcParams.update({"font.family": "serif"})
    
    # Create figure
    plt.figure(figsize=(8, 5), dpi=300)
    ax = sns.barplot(x='Model', y=metric, data=df, palette='Blues_d', edgecolor='black')

    # Annotate bars
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.3f}', 
                    (p.get_x() + p.get_width() / 2., p.get_height() + 0.005), 
                    ha='center', va='bottom', fontsize=11, fontweight='bold')

    # Labels
    metric_label = metric.replace("_", " ").title()
    plt.title(f'Model {metric_label} Comparison', fontsize=16, fontweight='bold')
    plt.ylabel(metric_label, fontsize=14, fontweight='bold')
    plt.xlabel('')
    plt.ylim(0.8, 1.02)
    plt.xticks(rotation=10)
    plt.tight_layout()
    plt.show()
```

```python
plot_model_metric_comparison("accuracy")
```

```python
plot_model_metric_comparison("precision")
```

```python
plot_model_metric_comparison("recall")
```

```python
plot_model_metric_comparison("f1_score")
```

```python
plot_model_metric_comparison("auc_roc")
```
